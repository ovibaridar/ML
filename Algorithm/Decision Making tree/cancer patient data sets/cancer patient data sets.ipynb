```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
path = 'https://raw.githubusercontent.com/ovibaridar/Data_sets/main/cancer%20patient%20data%20sets.csv'
```

```python
data = pd.read_csv(path)
```

```python
data.head()
```

```python
data = data.drop(['index','Patient Id'], axis=1)
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data.columns
```

```python
data['Level'] = data['Level'].replace({'Low': 1, 'Medium': 2, 'High': 3})
```

```python
data['Level'].value_counts()
```

# grafical view 

```python
sn.countplot(data= data ,x='Level')
```

```python
plt.figure(figsize=(10,10))
sn.heatmap(data.corr() ,annot = True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x= data.drop('Level', axis=1)
y= data.Level
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = .30,random_state = 42)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
plt.figure(figsize=(10,10))
sn.heatmap(xtrain.corr() ,annot = True)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(xtrain,ytrain)
```

```python
perd = dtc.predict(xtest)
perd
```

```python
dtc.score(xtrain,ytrain)
```

```python
dtc.score(xtest,ytest)
```

```python
from sklearn.metrics import r2_score ,accuracy_score 
```

```python
accuracy_score(ytest,perd)
```
